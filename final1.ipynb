```python
from bs4 import BeautifulSoup     
from selenium import webdriver
import time
import sys
import pandas as pd
import xlwt
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
path = "/Users/jun-mac/py_temp/chromedriver.exec"
driver = webdriver.Chrome(path)
driver.get("https://www.g2b.go.kr/index.jsp")

query_txt = input('1.공고문으로 검색할 키워드는 무엇입니까?:')
start=input('2.조회 시작일자 입력(예:2023/05/05):')
end=input('3.조회 종료일자 입력(예:2023/06/01):')
f_name = input('4.txt파일로 저장할 위치와 이름을 입력하시오:(예:/Users/jun-mac/final/final1.txt):')
fx_name = input('5.xls파일로 저장할 위치와 이름을 입력하시오:(예:/Users/jun-mac/final/final1.xls):')
#키워드 등록
driver.find_element(By.ID,'bidNm').click()
element =driver.find_element(By.ID,'bidNm')
element.send_keys(query_txt)
#시작일자
driver.find_element(By.ID,'fromBidDt').click()
driver.find_element(By.ID,'fromBidDt').send_keys(start)
#종료일자
driver.find_element(By.ID,'toBidDt').click()
driver.find_element(By.ID,'toBidDt').send_keys(end)
driver.find_element(By.CLASS_NAME, 'btn_dark').click()

time.sleep(2)
#frame이동 **이동 없이 접근 불가능
driver.switch_to.frame('sub')
driver.switch_to.frame('main')
temp =driver.find_element(By.CLASS_NAME, 'results')
div_list = temp.find_elements(By.TAG_NAME,'div')
#print(temp.text)
#div표 정보 가져오기#링크도 같이
results = []
for div in div_list:
    results.append(div.text)
    a_tags = div.find_elements(By.TAG_NAME, 'a')
    if a_tags:
        for a_tag in a_tags:
            link = a_tag.get_attribute('href')
            results.append(link)
            
temp_0=[]#업무
temp_1=[]#공고 번호-차수
temp_2=[]#공고 번호_상세
temp_3=[]#분류
temp_4=[]#공고명
temp_5=[]#공고명 상세
temp_6=[]#공고기관
temp_7=[]#수요기관
temp_8=[]#계약방법
temp_9=[]#입력일시(입찰마감일시)
temp_10=[]#공동수급
temp_11=[]#투찰

for i in range(10):
    temp_0.append(results[i*12])
    temp_1.append(results[i*12+1])
    temp_2.append(results[i*12+2])
    temp_3.append(results[i*12+3])
    temp_4.append(results[i*12+4])
    temp_5.append(results[i*12+5])
    temp_6.append(results[i*12+6])
    temp_7.append(results[i*12+7])
    temp_8.append(results[i*12+8])
    temp_9.append(results[i*12+9])
    temp_10.append(results[i*12+10])
    temp_11.append(results[i*12+11])
#txt저장
orig_stdout = sys.stdout
f = open(f_name , 'a' , encoding='UTF-8')
sys.stdout = f
time.sleep(1)

for i in range(10):
    print(i+1,'번째 업무')
    print('1','업무:',temp_0[i])
    print('2','공고 번호-차수:',temp_1[i])
    print('3','공고 번호_상세:',temp_2[i])
    print('4','분류:',temp_3[i])
    print('5','공고명:',temp_4[i])
    print('6','공고명 상세:',temp_5[i])
    print('7','공고기관:',temp_6[i])
    print('8','수요기관:',temp_7[i])
    print('9','계약방법:',temp_8[i])
    print('10','입력일시(입찰마감일시):',temp_9[i])
    print('11','공동수급:',temp_10[i])
    print('12','투찰:',temp_11[i])
    print('———————————————————————')
    
sys.stdout = orig_stdout
f.close()
#엑셀 파일로 저장
xls_temp = pd.DataFrame()

xls_temp['업무']=temp_0
xls_temp['공고 번호-차수']=temp_1
xls_temp['공고 번호_상세']=temp_2
xls_temp['분류']=temp_3
xls_temp['공고명']=temp_4
xls_temp['공고명 상세']=temp_5
xls_temp['공고기관']=temp_6
xls_temp['수요기관']=temp_7
xls_temp['계약방법']=temp_8
xls_temp['입력일시(입찰마감일시)']=temp_9
xls_temp['공동수급']=temp_10
xls_temp['투찰']=temp_11

xls_temp.to_excel(fx_name)
```
